```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('https://static.junilearning.com/python_data_science/advertisements.csv')

display(df)
```

```python
#Then display a histogram of cost for each type of advertising. Which medium does the company spend the most money on currently?

sns.histplot(df["TV"], bins=45)

plt.xlabel("$ Thousands Spent On TV Ads")
plt.ylabel("Number of Campaigns")
```

```python
sns.histplot(df["Radio"], bins=45)

plt.xlabel("$ Thousands Spent On Radio Ads")
plt.ylabel("Number of Campaigns")
plt.xlim(0,300)
```

```python
sns.histplot(df["Newspaper"], bins=45)

plt.xlabel("$ Thousands Spent On Newspaper Ads")
plt.ylabel("Number of Campaigns")
plt.xlim(0,300)
```

I recommend increasing spend on TV advertising and decreasing spend on newspaper and radio advertising.

Pearson's Coefficient measures the strength of a linear correlation between two variables. It ranges between -1 and 1. -1 represents a perfect negative correlation. 1 represents a perfect positive correlation. 0 means there's no correlation at all. 

Using the `strength` function below, quantify the strength of each correlation. Do they seem to match your findings in the graphs?

Student Analysis:


```python
# Starter Code
from scipy import stats
def strength(variable1, variable2):
  return stats.pearsonr(variable1,variable2)[0]

print(strength(df['TV'],df['Sales']))
print(strength(df['Newspaper'],df['Sales']))
print(strength(df['Radio'],df['Sales']))
```
